```python
import pandas as pd
music = pd.read_csv("charts.csv")
```

```python
music.isnull().sum()
```

```python
music['streams'].median()
```

```python
music['streams'] = music['streams'].fillna(music['streams'].median())
music.isnull().sum()
```

```python
music.dropna(inplace=True)
music.isnull().sum()
```

```python
music.to_csv('music.csv', index=False)
```
