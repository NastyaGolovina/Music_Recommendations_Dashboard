```python
import pandas as pd

df = pd.read_csv("artists.csv")

chunk_size = 3600

for i in range(0, len(df), chunk_size):
    chunk = df[i:i+chunk_size]
    chunk.to_csv(f"artists_part_{i//chunk_size}.csv", index=False)
```
